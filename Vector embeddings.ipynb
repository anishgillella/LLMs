Semantic Search means not searching using the exact keyword but understanding the intent of the user queryand using the context to perform the search

- It uses embedding which is the numerical  representation of words in a high dimensional space using a vector.
- Looking at the embedding of the word you can figure out the context though they are the same words.


To store the millions of vectors, we store them in a traditional database. The challenge is to search through each of records to find similar embeddings

In this case we use database index which leverages hashing function which creates buckets of similar looking embeddings and the search function to find similar embeddings is much faster and this technique is called Locality Sensitive Heading.

Benefots of vectors database is that they help us in faster search and optimized storage

An embedding is data like words that has been converted into an array of numbers. The combination of vectors serve as a multi dimensional map to find similarities.

A vector database can be used for searching, clustering, classification and recommendation.

```python
// Open AI URL and Headers
const openAiHeaders = {
    'Content-type' : 'application/json',
    'Authorization' : 'Bearer sk-****************************'
}

async function createEmbedding(texttoEmbed) {
    let response = await fetch("https://api.openai.com/v1/embeddings",
    {
        method : 'POST',
        headers : 'openAiHeaders',
        body : JSON.stringify(
            {
                'model' : 'text-embedding-ada-002',
                'input' : texttoEmbed
            })
    })
    if (response.ok) {
        response.json().then(data => {
            console.log(data);
            return data;
        });
    }
}


createEmbedding('Hello World')
```
